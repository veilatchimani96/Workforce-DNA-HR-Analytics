# **Stage 2 - Data Cleaning and Pre-Processing**

## Handling Missing Values in Categorical and Numerical Columns

```python
# Numerical Column Imputation
df['Monthly Income'].fillna(df['Monthly Income'].mean(),inplace=True)
```

```python
# Categorical Column Imputation
df['Education']=df['Education'].fillna("Unknown")
df['Marital Status']=df['Marital Status'].fillna("Unknown")
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip()
```

```python
df
```

## Creating Derived Columns

```python
df['Monthly Income'].describe()
```

```python
def Income_Leval(income):
    if income <= 5000:
        return 'Low'
    elif income <= 8000:
        return 'Medium'
    else:
        return 'High'

df['Income_Leval'] = df['Monthly Income'].apply(Income_Leval)
```

```python
df
```

```python
df['Income_Leval'].value_counts()
```

```python
# Create the age group column
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, include_lowest=True)
```

```python
df
```

## Employee Data Filtering

```python
# 1. Employees Who Left the Organization (Attrition = Yes)
attrition_emp = df[df['Attrition'] == 'Yes']
attrition_emp.head()
```

```python
# 2. Employees Working Overtime
overtime_emp = df[df['Over Time'] == 'Yes']
overtime_emp.head()
```

#**Stage 3 - EDA and Visulalization**

##**Univariate Analysis**

```python
# Histogram Plot
plt.figure(figsize=(6,4))
sns.histplot(df['Monthly Income'], bins=20, kde=True)
plt.title("Monthly Income Distribution")
plt.xlabel("Monthly Income")
plt.ylabel("Number of Employees")
plt.show()
```

# Interpretation - Monthly Income Histogram
###**Histogram Plot** Most employees earn between **₹2,000 and ₹7,000** per month, while the number of employees **decreases as monthly income increases.** This indicates that **only a small proportion of employees receive higher salaries.**

```python
# Bar Plot
plt.figure(figsize=(6,4))
sns.countplot(
    x='Age Group',
    data=df,
    order=['18-25', '26-35', '36-45', '46-55', '56-65'],
    palette=['orange', 'blue', 'blue', 'green', 'red'])
plt.title("Employee Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.show()
```

# Interpretation – Age Group Bar Chart
###**Bar Plot** Most employees are in the **26–35** and **36–45** age groups. Fewer employees belong to the **18–25** and **56–65** age groups, indicating that the workforce is mainly composed of **Mid-career professionals**.

```python
# Pie Plot
plt.figure(figsize=(5,5))
df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['blue', 'orange'])
plt.title("Employee Attrition Distribution")
plt.ylabel("")
plt.show()
```

# Interpretation- Employee Attrition Distribution (Pie Plot)
### **The pie chart shows that 83.9% of employees continue to work in the organization**, while **only 16.1% have left**. This indicates a **high employee retention rate** and a **relatively low attrition rate.**

```python
# Gender Distribution - Count Plot
plt.figure(figsize=(5,4))
sns.countplot(x='Gender',data=df,palette={'Male':'navy', 'Female':'deeppink'} )
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Employees")
plt.show()
```

# Interpretation - Gender Distripution
###**Count Plot** The number of **male employees is higher than female employees** in the organization. This indicates that the workforce is **male-dominated**, with females representing a smaller proportion of the total employees.

##**Bivariate Analysis**

```python
# Department vs Monthly Income - Box Plot

plt.figure(figsize=(8,5))
sns.boxplot( data=df,x='Department', y='Monthly Income')
plt.title("Department-wise Monthly Income Distribution")
plt.xlabel("Department")
plt.ylabel("Monthly Income")
plt.show()
```

# Ineterpretation - Department wise Monthly Income Distripution
###**Box Plot** The **Sales department has the highest median monthly income,** while **R&D and HR have relatively lower median incomes.** The presence of several **outliers** in all departments indicates that a small number of employees earn significantly higher salaries than the majority.

```python
# Count Plot
plt.figure(figsize=(6,4))
sns.countplot(x='Over Time',hue='Attrition',data=df,palette=['blue', 'purple'])
plt.title("Over Time vs Employee Attrition")
plt.xlabel("Over Time")
plt.ylabel("Number of Employees")
plt.legend(title="Attrition")
plt.show()
```

# Interpretation - Over Time vs Employee Attrition
###**Count Plot** Employees who **do not work overtime have the lowest attrition,** while employees who work overtime show a relatively **higher attrition rate.** This suggests that **overtime may be associated with increased employee turnover.**

```python
# Violin Plot
plt.figure(figsize=(12,6))
sns.violinplot(x='Job Role',y='Monthly Income',data=df,palette='Set1')
plt.title("Monthly Income Distribution by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Monthly Income")
plt.xticks(rotation=45)
plt.show()
```

# Interpretation - Monthly Income Distribution by Job Role
###**Violin Plot** **Managers** and **Research Directors** have the **highest monthly incomes,** while **Research Scientists,** **Laboratory Technicians,** and **Sales Representatives** earn comparatively lower incomes. This indicates that **monthly income varies significantly across different job roles.**

##**Multivariate Analysis**

```python
# Scatter Plot
plt.figure(figsize=(8,5))
sns.scatterplot(x='Age',y='Monthly Income',hue='Gender',data=df,s=70)
plt.title("Age vs Monthly Income by Gender")
plt.xlabel("Age")
plt.ylabel("Monthly Income")
plt.legend(title="Gender")
plt.show()
```

# Interpretation - Age vs Monthly Income by Gende
###**Scatter Plot** Monthly income generally **increases with age.** Both **male and female employees show a similar income distribution,** indicating **no major gender-based difference in monthly income across different age groups.**

```python
# Heatmap
plt.figure(figsize=(6,4))
numerical_cols = [
    'Age',
    'Monthly Income',
    'Total Working Years',
    'Years At Company']
sns.heatmap(
    df[numerical_cols].corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f')
plt.title("Correlation Heatmap of Employee Features")
plt.show()
```

# Interpretation - Correlation Heatmap of Employee Features
###**Heatmap** The heatmap shows a **strong positive correlation between Monthly Income and Total Working Years (0.77),** indicating that employees with more work experience tend to earn higher salaries. **Age** also has a **moderate positive correlation** with both **Monthly Income** and T**otal Working Years,** while **Years at Company** shows a moderate positive relationship with these features.

```python
# Pair Plot with Attrition
sns.pairplot(
    df[numerical_cols + ['Attrition']],
    hue='Attrition',
    diag_kind='hist'
)
plt.show()
```

# Interpretation - Pair Plot
###**Pair Plot** The pair plot shows **positive relationships among Age, Monthly Income, Total Working Years, and Years at Company.** Employees with  **higher experience and age generally earn higher incomes,** while the distributions of employees with and without attrition largely overlap, indicating **no clear separation based on these numerical features alone.**
